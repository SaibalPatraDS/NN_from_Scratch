## Implementing `Dense Layer` from Scratch

```python
import numpy as np
from nnfs.datasets import spiral_data

import matplotlib.pyplot as plt
```

```python
## getting a sprial data
X,y = spiral_data(1000, classes=3)
plt.scatter(X[:,0], X[:,1], c=y, cmap='brg')
plt.show()
```

### Creating `Dense Layer` Class

```python
## implementing Dense Layer

class DenseLayer:
    def __init__(self, n_inputs, n_neurons):
        self.n_neurons = n_neurons
        self.n_inputs = n_inputs
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)  ## assigning the weights
        self.biases = np.zeros((1,n_neurons)) ## assigning biases
    
    def forward(self,inputs):
        self.outputs = np.dot(inputs, self.weights) + self.biases 
```

### Implementing the Class

```python
## creating dense layer
# x.shape, y.shape -- ((3000, 2), (3000,))

dense_layer_1 = DenseLayer(2,3)
## applying forward function
dense_layer_1.forward(X)
## Observing the final results 
dense_layer_1.outputs[:5]
```
